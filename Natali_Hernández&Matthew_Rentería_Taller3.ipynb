## **Treemap con Matriculados en educación superior** 

```python
import pandas as pd  
import numpy as np
import seaborn as sbs
import matplotlib.pyplot as plt
```

```python
Base2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MEN_MATRICULA_ESTADISTICA_ES.csv')
Base2.head()
```

```python
Faltantes=pd.DataFrame(Base2.isna().sum().sort_values(ascending=False),columns=["Nulos"])
```

```python
Faltantes["Proporción"]=Faltantes["Nulos"]/Base2.shape[0]
Faltantes
```

```python
Base_limpia=Base2.dropna()
```

```python
Base_limpia.dtypes
```

```python
Base_limpia.isna().sum()
```

```python
Base_uso= Base_limpia.loc[Base2["Año"]==2015,["Institución de Educación Superior (IES)","Departamento de domicilio de la IES","Total Matriculados","Núcleo Básico del Conocimiento (NBC)"]]
Base_uso.shape
```

```python
import plotly.express as px
```

```python
fig=px.treemap(Base_uso,path=["Departamento de domicilio de la IES","Institución de Educación Superior (IES)", "Núcleo Básico del Conocimiento (NBC)"],values="Total Matriculados",color_discrete_map='RdBu',
               title="Distribución de por departamentos de Colombia matriculados en Educación Superior (IES)")
fig.update_layout()
fig.show()
```
